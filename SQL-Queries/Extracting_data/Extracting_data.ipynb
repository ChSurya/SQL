# **Using Orderby Clause**

**Orderby - ascending order**

```python
query = """
SELECT product_id,product_name
FROM `scaler-sql-476312.farmers_market.product`
ORDER BY product_id
"""
df = client.query(query).to_dataframe()
df
```

**Orderby - descending order**

```python
query = """
SELECT product_id,product_name
FROM `scaler-sql-476312.farmers_market.product`
ORDER BY product_id DESC
"""
df = client.query(query).to_dataframe()
df
```

**Both Ascending Order and Descending Order by using Order by Clause **

```python
query = """
SELECT product_id,product_name
FROM `scaler-sql-476312.farmers_market.product`
ORDER BY product_category_id DESC,product_id
"""
df = client.query(query).to_dataframe()
df
```

# By Using Limit Clause

```python
query = """
SELECT *
FROM `scaler-sql-476312.farmers_market.customer_purchases`
ORDER BY market_date DESC, transaction_time DESC
LIMIT 10;
"""
df = client.query(query).to_dataframe()
df
```

# **NOTE - Without using Limit we cannot use OFFSET Clause **

# OFFSET Clause

```python
query = """
SELECT *
FROM `scaler-sql-476312.farmers_market.customer_purchases`
ORDER BY market_date DESC, transaction_time DESC
LIMIT
  2 OFFSET 1;
"""
df = client.query(query).to_dataframe()
df
```

# **Using Alias **

```python
query = """
SELECT *, cost_to_customer_per_qty * quantity AS Revenue
FROM `scaler-sql-476312.farmers_market.customer_purchases`;
"""
df = client.query(query).to_dataframe()
df
```

# Using round() Function

```python
query = """
SELECT *, round(cost_to_customer_per_qty * quantity,2) AS Revenue
FROM `scaler-sql-476312.farmers_market.customer_purchases`;
"""
df = client.query(query).to_dataframe()
df
```
